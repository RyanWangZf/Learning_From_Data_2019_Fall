```python
# -*- coding: utf-8 -*-
__author__ = "ZifengWang"
__mail__ = "wangzf18@mails.tsinghua.edu.cn"
import matplotlib.pyplot as plt
plt.style.use("ggplot")
import pdb
import numpy as np
np.random.seed(2019)
from sklearn.decomposition import PCA, KernelPCA
from sklearn.decomposition import FastICA
from sklearn.cross_decomposition import CCA
from sklearn import datasets

from utils import plot_demo_v1, HGR
```

```python
# synthetic data
X,y = datasets.make_moons(n_samples=100, noise=.05)
# X,y = datasets.make_circles(n_samples=100, factor=.3, noise=.05)
# X,y = datasets.make_blobs(n_samples=100)

x1 = X[:,0].reshape(-1,1)
x2 = X[:,1].reshape(-1,1)
reds = y == 0
blues = y == 1
```

```python
# PCA
pca = PCA()
X_pca = pca.fit_transform(X)
fx1 = X_pca[:,0]
gx2 = X_pca[:,1]
plot_demo_v1(x1,x2,fx1,gx2,reds,blues,"PCA")
```

```python
# Kernel PCA
kpca = KernelPCA(kernel="rbf", gamma=10)
X_kpca = kpca.fit_transform(X)
fx1 = X_kpca[:,0]
gx2 = X_kpca[:,1]
plot_demo_v1(x1,x2,fx1,gx2,reds,blues,"K-PCA")
```

```python
# ICA
ica = FastICA()
X_ica = ica.fit_transform(X)
fx1 = X_ica[:,0]
gx2 = X_ica[:,1]
plot_demo_v1(x1,x2,fx1,gx2,reds,blues,"ICA")
```

```python
# CCA
cca = CCA(n_components=1)
cca.fit(x1,x2)
X_cca = cca.transform(x1,x2)
fx1, gx2 = X_cca
plot_demo_v1(x1,x2,fx1,gx2,reds,blues,"CCA")
```

```python
# HGR
hgr = HGR(1,1)
hgr.fit(x1,x2,20000)
X_hgr = hgr.transform(x1,x2)
fx1 = X_hgr[0]
gx2 = X_hgr[1]
plot_demo_v1(x1,x2,fx1,gx2,reds,blues,"HGR")
```
